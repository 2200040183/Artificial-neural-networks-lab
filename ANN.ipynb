```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
scaler=StandardScaler()
X=scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)

```

```python
multi_layer=MLPClassifier(hidden_layer_sizes=(8,8,8),max_iter=100,random_state=25)
multi_layer.fit(X_train,y_train)
```

```python
predict=multi_layer.predict(X_test)
```

```python
print(predict)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(y_test,predict))
print(classification_report(y_test,predict))
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler=StandardScaler()
X=scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)

```

```python
multi_layer=MLPClassifier(hidden_layer_sizes=(5,5,5),max_iter=1000,random_state=42)
multi_layer.fit(X_train,y_train)
```

```python
y_predict=multi_layer.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def linear(x):
    return x
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
def bipolar(x):
  return 2 / (1 + np.exp(-2*x)) - 1
def tanh(x):
  return np.tanh(x)
def relu(x):
  return np.maximum(0,x)
def leaky_relu(x):
  return np.maximum(0.01*x,x)
def softplus(x):
  return np.log(1 + np.exp(x))
x=np.linspace(-10,10,50)
plt.figure(figsize=(15,15))
plt.plot(x,linear(x))
plt.title('Linear')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,sigmoid(x))
plt.title('Sigmoid')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,bipolar(x))
plt.title('Bipolar')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,tanh(x))
plt.title('Tanh')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,relu(x))
plt.title('ReLU')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,leaky_relu(x))
plt.title('Leaky ReLU')
plt.show()
plt.figure(figsize=(15,15))
plt.plot(x,softplus(x))
plt.title('Softplus')
plt.show()




```
